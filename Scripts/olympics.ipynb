```python
def asciiConvert(attr):
    attr = str(attr)
    asciiSum = 0
    for c in attr: 
        asciiSum += (ord(c))
    return asciiSum
```

```python
def hashVal(attr, k):
    asciiSum = asciiConvert(attr)
    return asciiSum%k
```

```python
def partition(data, k, attr = None):
    d = {}
    for row in data.index:
        attrVal = data.loc[row][attr]
        h = hashVal(attrVal, k)
        if h not in d.keys():
            d[h] = []
        d[h].append(data.loc[row].to_json())
    return d
```
